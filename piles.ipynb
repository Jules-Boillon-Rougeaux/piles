# Pile et POO
Les représentations précédentes utilisent un type existant, le type ```list```.
Dans la programmation orientée objet, on définit une structure nouvelle.  
Nous considérons des piles à capacité non bornée.  
On définit une classe ```Pile``` avec : 

des **attributs**:
* taille : un entier (nombre d'éléments contenus dans la pile);
* éléments (la pile) : une liste de longueur taille (type list); 

des **méthodes**:
* procédure ```init```, le constructeur (fixe la taille à 0 et crée une liste vide);
* procédure ```empiler```;
* fonction ```depiler```;
* fonction ```pile_vide```;
* fonction ```sommet```;
* des méthodes d'affichages.  

Ecrire le programme de cette nouvelle structure puis la tester.

````python
class Pile():
    def init():
        taille=0
        éléments=[]
        return éléments
    
    def empiler(pile,e):
        pile.append(e)
        return pile
    
    def depiler(pile):
        pile.pop(-1)
        return pile
    
    def pile_vide(pile):
        return pile==[]
    
    def sommet(pile):
        return pile[-1]
    
    def affichage(pile):
        pile_str=""
        for i in range(len(pile)-1):
            pile_str=pile_str+str(pile[i])+" ; "
        pile_str=pile_str+str(pile[i+1])
        return pile_str
````

````python
p = Pile.init()
````
